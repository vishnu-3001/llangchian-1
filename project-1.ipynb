```python
from langchain import OpenAI
from langchain.chains import RetrievalQA
```

```python
import os
```

```python
llm=OpenAI(temperature=0.9,max_tokens=500)
loader = UnstructuredURLLoader(
    urls = [
        "https://www.moneycontrol.com/news/business/banks/hdfc-bank-re-appoints-sanmoy-chakrabarti-as-chief-risk-officer-11259771.html",
        "https://www.moneycontrol.com/news/business/markets/market-corrects-post-rbi-ups-inflation-forecast-icrr-bet-on-these-top-10-rate-sensitive-stocks-ideas-11142611.html"
    ]
)
data=loader.load()
len(data)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
texts=text_splitter.split_documents(data)
len(texts)
```

```python
from langchain.vectorstores import FAISS
```

```python
!pip install tiktoken
```

```python
from langchain.embeddings import OpenAIEmbeddings
embeddings=OpenAIEmbeddings()
vector_index=FAISS.from_documents(texts,embeddings)
```

```python
from langchain.chains import RetrievalQAWithSourcesChain
```

```python
chain = RetrievalQAWithSourcesChain.from_chain_type(
    llm=llm,
    chain_type="stuff",  # Specify chain type
    retriever=vector_index.as_retriever(),
    return_source_documents=True  # Enable returning source documents
)
```

```python
import langchain
query="where does hdfc cro studied his degree"
# langchain.debug = True

# Input data for the chain
input_data = {"question": query}

# Execute the chain and get results
result = chain(input_data)

# Print results
print(result)
```
